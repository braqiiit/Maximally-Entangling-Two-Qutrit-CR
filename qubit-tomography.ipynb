```python
import numpy as np

from statetomography import QubitStateTomography
from processtomography import QubitProcessTomography
```

### Performing Qubit State Tomograph of the state $\lvert \psi \rangle = \frac{1}{2}\lvert0\rangle + \frac{\sqrt{3}}{2}\lvert1\rangle$.

```python
state = np.array([1/2, np.sqrt(3)/2])
qst = QubitStateTomography(state)
```

```python
coeffs = qst.run()
print(f"The given state is rho = {coeffs[0]}*I + {coeffs[1]}*X + {coeffs[2]}*Y + {coeffs[3]}*Z")
```

### Performing Qubit Process Tomography of the gate $Y = \begin{bmatrix}0 & -i\\ i & 0 \end{bmatrix}$.

```python
gate = np.array([[np.sqrt(3)/2,1/2], [1/2, -np.sqrt(3)/2]])
qpt = QubitProcessTomography(gate)
```

```python
chi_matrix = qpt.run()
```

```python
qpt.plot_chi()
```

```python
np.trace(np.matmul( np.conjugate(chi_matrix.transpose()), chi_matrix))
```
